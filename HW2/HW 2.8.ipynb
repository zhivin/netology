```python
import random
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
#Задние №1 Рассчитать и визуализировать корреляционную матрицу для переменных
```

```python
data = pd.read_csv('wine_data.csv')
corr = data.corr()

from pylab import rcParams
rcParams['figure.figsize'] = 10, 10

sns.heatmap(corr, annot=True,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
#Задание №2 Визуализировать взаимосвязи между 3-4 переменным
```

```python
# взаимосвязь уровня аклоголя и насыщенности цвета

data = pd.read_csv('wine_data.csv')
data = data.sort_values(by = 'Color intensity', ascending = False)

alc = [i for i in data['Alcohol']]
color = [i for i in data['Color intensity']]

from pylab import rcParams
rcParams['figure.figsize'] = 10, 10

plt.plot(alc, color)

plt.xlabel('уровень алкоголя')
plt.ylabel('насыщенность цвета')
```

```python
# взаимосвязь уровня золы и магния

data = pd.read_csv('wine_data.csv')
data = data.sort_values(by = 'Magnesium', ascending = False)

mag = [i for i in data['Magnesium']]
ash = [i for i in data['Ash']]

from pylab import rcParams
rcParams['figure.figsize'] = 10, 10

plt.plot(mag, ash)

plt.xlabel('уровень золы')
plt.ylabel('уровень магния')
```

```python
# взаимосвязь уровня фенолов и флавонидов

data = pd.read_csv('wine_data.csv')
data = data.sort_values(by = 'Flavanoids', ascending = False)

phenols = [i for i in data['Total phenols']]
flav = [i for i in data['Flavanoids']]

from pylab import rcParams
rcParams['figure.figsize'] = 10, 10

plt.plot(flav, phenols)

plt.xlabel('уровень флавонидов')
plt.ylabel('уровень фенолов')
```

```python
#Задание №3 Рассчитать t-test статистики для 3-4 переменных
```

```python
cls1 = data[data['Class'] == 1]['Alcohol']
cls2 = data[data['Class'] == 2]['Alcohol']

stats.ttest_ind(cls1, cls2) 
```

```python
stats.ttest_rel(data['Flavanoids'], data['Total phenols'])  
```

```python
stats.ttest_rel(data['Ash'], data['Magnesium'])  
```
